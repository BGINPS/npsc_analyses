```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, TensorDataset, Dataset
from tqdm import tqdm

#自定义数据集类
class TimeSeriesDataset(Dataset):
    def __init__(self, sequences):
        self.sequences = sequences
    
    def __len__(self):
        return len(self.sequences)
    
    def __getitem__(self, idx):
        sequence = self.sequences[idx]
        return torch.tensor(sequence, dtype=torch.float32)



class TimeSeriesAutoencoder(nn.Module):
    """增强型时间序列自编码器"""
    def __init__(self, input_size, encoding_dim):
        super().__init__()
        
        # 编码器：使用1D卷积
        self.encoder = nn.Sequential(
            nn.Conv1d(1, 16, kernel_size=5, stride=2, padding=2),  # [batch, 16, seq_len/2]
            nn.ReLU(),
            nn.Conv1d(16, 32, kernel_size=5, stride=2, padding=2),  # [batch, 32, seq_len/4]
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(32 * (input_size // 4), 128),
            nn.ReLU(),
            nn.Linear(128, encoding_dim)
        )
        
        # 解码器：使用转置卷积
        self.decoder = nn.Sequential(
            nn.Linear(encoding_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 32 * (input_size // 4)),
            nn.Unflatten(1, (32, input_size // 4)),
            nn.ConvTranspose1d(32, 16, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose1d(16, 1, kernel_size=5, stride=2, padding=2, output_padding=1),
            nn.Sigmoid()  # 确保输出在合理范围
        )
    
    def forward(self, x):
        # 添加通道维度 [batch, 1, seq_len]
        x = x.unsqueeze(1) if x.dim() == 2 else x
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded.squeeze(1)  # 移除通道维度


# 生成模拟数据
def generate_synthetic_data(num_sequences=1000, seq_length=100, anomaly_ratio=0.1):
    """
    生成合成时间序列数据
    :param num_sequences: 序列数量
    :param seq_length: 序列长度
    :param anomaly_ratio: 异常序列比例
    :return: 时间序列数组
    """
    # 生成基础趋势 (正弦波)
    base_trend = np.sin(np.linspace(0, 4 * np.pi, seq_length))
    
    sequences = []
    labels = []
    
    for i in range(num_sequences):
        # 添加随机噪声
        noise = np.random.normal(0, 0.2, seq_length)
        
        # 创建正常序列 (基于基础趋势)
        if np.random.rand() > anomaly_ratio:
            # 添加随机振幅和相位变化
            amplitude = 0.5 + np.random.rand() * 0.5
            phase_shift = np.random.randint(0, 20)
            sequence = amplitude * np.roll(base_trend, phase_shift) + noise
            sequences.append(sequence)
            labels.append(0)  # 正常序列
        else:
            # 创建不一致序列
            if np.random.rand() < 0.5:
                # 类型1: 不同趋势
                anomaly_trend = np.linspace(0, 1, seq_length) * np.sin(np.linspace(0, 8 * np.pi, seq_length))
                sequence = anomaly_trend + noise
            else:
                # 类型2: 随机序列
                sequence = np.random.randn(seq_length)
            sequences.append(sequence)
            labels.append(1)  # 异常序列
    
    return np.array(sequences), np.array(labels)

# 数据预处理
def preprocess_data(sequences):
    """
    标准化时间序列数据
    :param sequences: 原始时间序列数组
    :return: 标准化后的序列
    """
    scaler = StandardScaler()
    normalized_sequences = []
    
    for seq in sequences:
        normalized = scaler.fit_transform(seq.reshape(-1, 1)).flatten()
        normalized_sequences.append(normalized)
    
    return np.array(normalized_sequences)

# 训练自编码器
def train_autoencoder(model, train_loader, val_loader, epochs=100, lr=0.001):
    """
    训练自编码器模型
    :param model: 自编码器模型
    :param train_loader: 训练数据加载器
    :param val_loader: 验证数据加载器
    :param epochs: 训练轮数
    :param lr: 学习率
    :return: 训练好的模型，训练损失历史
    """
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    
    train_loss_history = []
    val_loss_history = []
    
    best_val_loss = float('inf')
    
    for epoch in range(epochs):
        # 训练阶段
        model.train()
        train_loss = 0.0
        
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} - Training"):
            sequences = batch.to(device)
            
            # 前向传播
            reconstructed = model(sequences)
            loss = criterion(reconstructed, sequences)
            
            # 反向传播
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item() * sequences.size(0)
        
        train_loss /= len(train_loader.dataset)
        train_loss_history.append(train_loss)
        
        # 验证阶段
        model.eval()
        val_loss = 0.0
        
        with torch.no_grad():
            for batch in val_loader:
                sequences = batch.to(device)
                reconstructed = model(sequences)
                loss = criterion(reconstructed, sequences)
                val_loss += loss.item() * sequences.size(0)
        
        val_loss /= len(val_loader.dataset)
        val_loss_history.append(val_loss)
        
        print(f"Epoch {epoch+1}/{epochs} - Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}")
        
        # 保存最佳模型
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), "best_autoencoder.pth")
    
    # 加载最佳模型
    model.load_state_dict(torch.load("best_autoencoder.pth"))
    
    return model, train_loss_history, val_loss_history

# 识别不一致序列
def detect_inconsistent_sequences(model, sequences, threshold_factor=2.0):
    """
    使用训练好的自编码器检测不一致序列
    :param model: 训练好的自编码器
    :param sequences: 时间序列数组
    :param threshold_factor: 阈值倍数
    :return: 不一致序列索引，重构误差
    """
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()
    
    # 创建数据加载器
    dataset = TimeSeriesDataset(sequences)
    loader = DataLoader(dataset, batch_size=64, shuffle=False)
    
    reconstruction_errors = []
    
    with torch.no_grad():
        for batch in loader:
            sequences_batch = batch.to(device)
            reconstructed = model(sequences_batch)
            
            # 计算每个序列的MSE
            mse = nn.functional.mse_loss(reconstructed, sequences_batch, reduction='none')
            batch_errors = mse.mean(dim=1).cpu().numpy()
            reconstruction_errors.extend(batch_errors)
    
    reconstruction_errors = np.array(reconstruction_errors)
    
    # 计算阈值
    median_error = np.median(reconstruction_errors)
    mad = np.median(np.abs(reconstruction_errors - median_error))
    threshold = median_error + threshold_factor * mad
    
    # 识别不一致序列
    inconsistent_indices = np.where(reconstruction_errors > threshold)[0]
    
    return inconsistent_indices, reconstruction_errors

# 可视化结果
def visualize_results(sequences, labels, reconstruction_errors, inconsistent_indices, threshold):
    """
    可视化结果
    :param sequences: 所有序列
    :param labels: 真实标签 (0=正常, 1=异常)
    :param reconstruction_errors: 重构误差
    :param inconsistent_indices: 检测到的不一致序列索引
    :param threshold: 检测阈值
    """
    # 1. 绘制重构误差分布
    plt.figure(figsize=(12, 6))
    
    # 正常序列的误差
    normal_errors = reconstruction_errors[labels == 0]
    # 异常序列的误差
    abnormal_errors = reconstruction_errors[labels == 1]
    
    plt.hist(normal_errors, bins=50, alpha=0.7, label='zhengchang')
    plt.hist(abnormal_errors, bins=50, alpha=0.7, label='yichang')
    plt.axvline(threshold, color='r', linestyle='--', label=f'阈值 ({threshold:.4f})')
    plt.title('chonguowucha')
    plt.xlabel('chonguowucha')
    plt.ylabel('pinshu')
    plt.legend()
    plt.show()
    
    # 2. 绘制ROC曲线
    from sklearn.metrics import roc_curve, auc, confusion_matrix
    
    fpr, tpr, _ = roc_curve(labels, reconstruction_errors)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(10, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC曲线 (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('假阳性率')
    plt.ylabel('真阳性率')
    plt.title('接收者操作特征曲线')
    plt.legend(loc="lower right")
    plt.show()
    
    # 3. 混淆矩阵
    predicted_labels = reconstruction_errors > threshold
    cm = confusion_matrix(labels, predicted_labels)
    
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('混淆矩阵')
    plt.colorbar()
    tick_marks = np.arange(2)
    plt.xticks(tick_marks, ['zhengchan', 'yichan'])
    plt.yticks(tick_marks, ['zhenchang', 'yichan'])
    
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], 'd'),
                     ha="center", va="center",
                     color="white" if cm[i, j] > thresh else "black")
    
    plt.ylabel('zhenshibiaoqian')
    plt.xlabel('yucebiaoqian')
    plt.show()
    
    # 4. 可视化示例序列
    fig, axes = plt.subplots(3, 3, figsize=(15, 10))
    fig.suptitle('序列示例：正常 vs 异常')
    
    # 随机选择一些正常序列
    normal_indices = np.where(labels == 0)[0]
    selected_normal = np.random.choice(normal_indices, 3, replace=False)
    
    # 随机选择一些异常序列
    abnormal_indices = np.where(labels == 1)[0]
    selected_abnormal = np.random.choice(abnormal_indices, 3, replace=False)
    
    # 随机选择一些检测到的不一致序列
    selected_detected = np.random.choice(inconsistent_indices, 3, replace=False)
    
    for i, idx in enumerate(selected_normal):
        ax = axes[0, i]
        ax.plot(sequences[idx])
        ax.set_title(f'zhengchan {idx}\nwuchan: {reconstruction_errors[idx]:.4f}')
    
    for i, idx in enumerate(selected_abnormal):
        ax = axes[1, i]
        ax.plot(sequences[idx])
        ax.set_title(f'yichan {idx}\nwuchan: {reconstruction_errors[idx]:.4f}')
    
    for i, idx in enumerate(selected_detected):
        ax = axes[2, i]
        ax.plot(sequences[idx])
        detection_type = "yichan" if labels[idx] == 1 else "zhengchan"
        ax.set_title(f'检测为不一致 ({detection_type})\n误差: {reconstruction_errors[idx]:.4f}')
    
    plt.tight_layout()
    plt.show()

# 主函数
def main():
    # 1. 生成和预处理数据
    sequences, labels = generate_synthetic_data(
        num_sequences=1000, 
        seq_length=100, 
        anomaly_ratio=0.15
    )
    normalized_sequences = preprocess_data(sequences)
    
    # 2. 划分训练集和测试集
    train_sequences, test_sequences, train_labels, test_labels = train_test_split(
        normalized_sequences, labels, test_size=0.3, random_state=42
    )
    
    # 3. 创建数据加载器
    train_dataset = TimeSeriesDataset(train_sequences)
    test_dataset = TimeSeriesDataset(test_sequences)
    
    batch_size = 64
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    
    # 4. 初始化模型
    input_size = sequences.shape[1]
    encoding_dim = 16  # 编码维度
    model = TimeSeriesAutoencoder(input_size, encoding_dim)
    
    # 5. 训练模型
    trained_model, train_loss, val_loss = train_autoencoder(
        model, train_loader, val_loader, epochs=100, lr=0.001
    )
    
    # 6. 绘制训练损失
    plt.figure(figsize=(10, 6))
    plt.plot(train_loss, label='训练损失')
    plt.plot(val_loss, label='验证损失')
    plt.title('训练和验证损失')
    plt.xlabel('Epoch')
    plt.ylabel('MSE损失')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # 7. 在整个数据集上检测不一致序列
    inconsistent_indices, reconstruction_errors = detect_inconsistent_sequences(
        trained_model, normalized_sequences, threshold_factor=2.0
    )
    
    # 计算阈值（用于可视化）
    median_error = np.median(reconstruction_errors)
    mad = np.median(np.abs(reconstruction_errors - median_error))
    threshold = median_error + 2.0 * mad
    
    print(f"检测到 {len(inconsistent_indices)} 个不一致序列")
    print(f"真实异常数量: {sum(labels)}")
    
    # 8. 可视化结果
    visualize_results(
        sequences, 
        labels, 
        reconstruction_errors, 
        inconsistent_indices,
        threshold
    )
    
    # 9. 返回一致序列
    consistent_indices = [i for i in range(len(sequences)) if i not in inconsistent_indices]
    consistent_sequences = sequences[consistent_indices]
    
    print(f"保留的一致序列数量: {len(consistent_sequences)}")
    
    return consistent_sequences

if __name__ == "__main__":
    consistent_sequences = main()
```
