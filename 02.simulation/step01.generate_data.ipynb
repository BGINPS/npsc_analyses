```python
npsc.generate_time_series_as_obj(situation='situation1', save_name='sim1', save_dir='../../03.results/simulation/sim1/')
```

```python
sim1_obj = nps.io.read_pickle('../../03.results/simulation/sim1/sim1.pkl')
sim1_target_obj = nps.tl.extract_reads_with_labels(sim1_obj, labels=['target'])
sim1_noise_obj = nps.tl.extract_reads_with_labels(sim1_obj, labels=['noise'])
```

```python
nps.pl.draw_one_read_with_window(sim1_noise_obj, list(sim1_noise_obj.keys())[1])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(sim1_noise_obj[list(sim1_noise_obj.keys())[0]]['signal'], c='blue', label='noise')
plt.plot(sim1_noise_obj[list(sim1_noise_obj.keys())[1]]['signal'], c='blue', alpha=0.5)
plt.plot(sim1_target_obj[list(sim1_target_obj.keys())[0]]['signal'], c='red', label='target')
plt.plot(sim1_target_obj[list(sim1_target_obj.keys())[1]]['signal'], c='red', alpha=0.5)
plt.legend()
plt.savefig("../../05.figures/simulation/sim1_case.pdf")
```
