```python
all_peps = ['hp1_1', 'hp1_2', 'hp1_3', 'hp1_4', 'hp1_5', 'hp1_6', 'hp1_7', 'hp1_8', 'hp1_9',
            'hp2_1', 'hp2_2', 'hp2_3', 'hp2_4', 'hp2_5', 'hp2_6',]
            #'hp3_2', 'hp3_3', 'hp3_4', 'hp3_5', 'hp3_6', 'hp3_7', 'hp3_9', 'hp3_10', 'hp3_11']
```

```python
stat_df = []
for pep in all_peps:
    clean_80_obj = nps.io.read_pickle(f"../../../03.results/classification_on_clean_data/hp12/{pep}/{pep}_valid80_clean_obj.pkl")
    clean_20_obj = nps.io.read_pickle(f"../../../03.results/classification_on_clean_data/hp12/{pep}/{pep}_valid20_clean_obj.pkl")
    stat_df.append([pep, len(clean_80_obj), len(clean_20_obj)])
stat_df = pd.DataFrame(stat_df, columns=['pep', 'v80_clean', 'v20_clean'])
stat_df = stat_df.set_index('pep')
stat_df
```

```python
stat_df.sort_values(by='v80_clean')
```

```python
valid_stat_df = pd.read_csv("../../../../opo/02.analyses/02.hp123/03.tables/combine_valid_read_stat.csv", index_col=0)
```

```python
valid_stat_df = valid_stat_df[['valid80', 'valid20']]
```

```python
all_stat_df = pd.concat([valid_stat_df, stat_df], axis=1)
all_stat_df
```

```python
all_stat_df['v80_clean_percent'] = all_stat_df['v80_clean'] / all_stat_df['valid80']
all_stat_df['v20_clean_percent'] = all_stat_df['v20_clean'] / all_stat_df['valid20']
```

```python
all_stat_df.to_csv('../../../04.tables/classification/hp/filter_stat.csv')
all_stat_df
```
